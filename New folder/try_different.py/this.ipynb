```python
import json

# Load the data from the first file
with open('file1.json', 'r') as file:
    data1 = json.load(file)

# Load the data from the second file
with open('file2.json', 'r') as file:
    data2 = json.load(file)

# Concatenate the data
# Assuming you want to add data2 into the array in data1
data1['result']['data']['json'] += data2

# Save the concatenated data to a new JSON file
with open('output.json', 'w') as file:
    json.dump(data1, file, indent=4)

print("Data has been concatenated and saved to 'output.json'")
```
